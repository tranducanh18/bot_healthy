```python
#!pip install -q transformers torch

from transformers import pipeline

health_agent = pipeline(
    "text2text-generation",
    model="google/flan-t5-large",
)

def ai_health_agent(user_question, max_new_tokens=250, temperature=0.7):
#user_question: cau hoi ban hoi AI
    prompt = f"""
You are a professional and friendly medical AI assistant.
Answer the user's question clearly, accurately, and concisely.
Provide suggestions, precautions, or explanations if relevant.
Do not give false information. Indicate if you are not sure.
User question: {user_question}
AI answer:
"""
    result = health_agent(
        prompt,
        max_new_tokens=max_new_tokens,
        do_sample=True,
        temperature=temperature
    )
    return result[0]["generated_text"]

print("=== AI Health Agent ===")
while True:
    question = input("You: ")
    if question.lower() in ["exit", "quit"]:
        print("AI: Goodbye! Take care of your health! 🏥")
        break
    answer = ai_health_agent(question)
    print(f"AI: {answer}")
```
